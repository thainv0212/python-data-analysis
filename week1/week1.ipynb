## Reading and Writing CSV files

```python
import csv
%precision 2
with open('mpg.csv') as csvfile:
    mpg = list(csv.DictReader(csvfile))
mpg
```

```python
len(mpg)
```

```python
mpg[0].keys()
```

```python
sum(float(d['displ']) for d in mpg)/len(mpg)
```

```python
cylinders = set(d['cyl']for d in mpg)
cylinders
```

```python
CtyMpgByCyl = []
for c in cylinders:
    summpg = 0
    sumcount = 0
    for d in mpg:
        if d['cyl'] == c:
            summpg += float(d['cty'])
            sumcount +=1
    CtyMpgByCyl.append((c, summpg / sumcount))
CtyMpgByCyl.sort(key= lambda x: x[1])
CtyMpgByCyl
```

```python
vehicleclass = set(d['class'] for d in mpg)
vehicleclass
```

** Iterating over Arrays **

```python
o = np.random.randint(0,10,(3,4))
```

```python
for row1, row2 in zip(o, o2):
    print('{} + {} = {}'.format(row1, row2, row1 + row2))
```
